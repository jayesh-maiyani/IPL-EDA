```python
sns.set()
```

```python
def bargraph(data, x, y, dpi):
    plt.figure(figsize=(12,8), dpi = dpi)

    plt.bar(data = data, x = x, height = y, color=(0.5, 0.5, 0.5, 0.5),  edgecolor='blue')

    plt.xticks(rotation=90)
#     plt.legend(bbox_to_anchor=(1, 1), loc=2)

    plt.show()
```

```python
# players with most 'player of the matches' award 
# greater than 10 times

bargraph(y, 'player_of_match', 'frequency', 150)
```

```python
best_team_of_season = seasons.winner.describe().drop(columns = ['count', 'unique'])
```

```python
best_team_of_season.reset_index(inplace = True)
```

```python
best_team_of_season
```

```python
plt.figure(figsize = (8,5), dpi = 100)
teams = best_team_of_season['top']
plt.xticks(rotation = 90)
ax = plt.gca()
plt.yticks(range(15))
pps = ax.bar(data=best_team_of_season, height='freq', x = 'Season', 
             color = 'black', alpha = 0.35, edgecolor = 'blue')
i = 0
for p in pps:
    height = p.get_height()
    ax.text(x=p.get_x() + p.get_width() / 2, 
            y=(height)//2 - 2, s="{} ({})".format(teams[i], height), 
            ha='center', fontsize = 'small', rotation = 90, animated = False, color = 'white')
    i += 1

plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (8,5), dpi = 100)
plt.xticks(rotation = 90)
ax = plt.gca()
sns.set_theme(style="whitegrid")
sns.barplot(data=best_team_of_season, y ='freq', x = 'Season')
i = 0
for p in pps:
    height = p.get_height()
    ax.text(x=p.get_x() + p.get_width() / 2, 
            y=(height)//2 - 2, s="{} ({})".format(teams[i], height), 
            ha='center', fontsize = 'medium', rotation = 90, animated = False, color = 'white')
    i += 1
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (8,5), dpi = 100)
plt.xticks(rotation = 90)
ax = plt.gca()
sns.set_theme(style="whitegrid")
sns.barplot(data=best_team_of_season, x ='freq', y = 'Season')
i = 0
for p in pps:
    height = p.get_height()
    ax.text(y=p.get_x() + .2 + p.get_width() / 2, 
            x=(height)//2, s="{} ({})".format(teams[i], height), 
            ha='center', fontsize = 'medium', animated = False, color = 'white')
    i += 1
plt.xticks(rotation = 90)
plt.show()
```

```python
## all wins
all_wins = matches_df[['Season', 'team-1', 'team-2', 'winner', 'win_by_runs', 'win_by_wickets']].sort_values('win_by_runs', ascending = False)
```

```python
## wins by chasing
batting_wins = all_wins[all_wins['win_by_wickets'] != 0].sort_values('win_by_wickets', ascending = False, ignore_index = True).drop(columns = ['win_by_runs'])
```

```python
## wins by defending
bowling_wins = all_wins[all_wins['win_by_runs'] != 0].sort_values('win_by_runs', ascending = False, ignore_index = True).drop(columns = ['win_by_wickets'])
```

```python
bowling_wins_groups = bowling_wins.dropna().groupby('winner')
```

```python
bowling_wins_groups.value_counts()
```
